```python
import matplotlib.pyplot as plt
import pandas as pd
import xarray as xr
```

```python
file = 'us-counties.csv'
counties = pd.read_csv(file)
counties.head()
```

```python
file = 'us-states.csv'
states = pd.read_csv(file)
states.head()
```

```python
state_totals = states.groupby('state').sum()
state_totals.head()
```

```python
state_totals.index
```

```python
plt.figure(figsize=(9,4), dpi=150)
plt.bar(state_totals.index, state_totals.cases, align='center', width=1)
plt.gca().tick_params(axis='x', rotation=90)
# plt.yscale('log')
```
